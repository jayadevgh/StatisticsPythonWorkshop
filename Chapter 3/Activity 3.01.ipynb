```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data3.csv')
```

```python
df.head()
```

```python
df = df.replace('?', np.nan)
```

```python
age = ['12t21', '12t29', '16t24', '65up']

for group in age:
    df['ageCount' + group] = (df['population'] * df['agePct' + group]).astype(int)
    
df[['population'] + ['agePct' + group for group in age] + ['ageCount' + group for group in age]].head()
```

```python
state_group = df.groupby('state')
state_group.sum()[['ageCount' + group for group in age]]
```

```python
df.describe()
```

```python
crime_df = df[['burglPerPop','larcPerPop','autoTheftPerPop','arsonsPerPop','nonViolPerPop']]

f, ax = plt.subplots(figsize=(13, 10))
sns.boxplot(data=  crime_df)
plt.show()
```

```python
feature_columns = ['PctPopUnderPov', 'PctLess9thGrade', 'PctUnemployed', 'ViolentCrimesPerPop', 'nonViolPerPop']
filtered_df = df[feature_columns]

f, ax = plt.subplots(figsize=(13, 10))
sns.heatmap(filtered_df.dropna().astype(float).corr(), center=0, annot=True)
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)

plt.show()
```
